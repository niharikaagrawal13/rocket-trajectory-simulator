# Rocket Trajectory Simulator

## Project Overview
This project involved developing a 3-DOF physics simulator from the ground up to analyze rocket performance. The core purpose was to investigate the complex interplay between design parameters (like thrust and mass) and their real-world effects on flight outcomes such as altitude, velocity, and structural loads.

The simulator solves the equations of motion using numerical integration (Euler's method), modeling forces including thrust, gravity, and aerodynamic drag.

```python
# This tests our essential libraries
import numpy as np
import matplotlib.pyplot as plt

print("✅ SUCCESS! All libraries are installed and working!")
print("NumPy version:", np.__version__)

# Let's make a simple rocket-related plot to prove it
time = np.linspace(0, 10, 100) # 10 seconds, 100 points
velocity = 50 * (1 - np.exp(-time / 2)) # Simulate a rocket accelerating

plt.figure(figsize=(10, 5))
plt.plot(time, velocity, 'r-', linewidth=2)
plt.title('Simulated Rocket Velocity vs. Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Velocity (m/s)')
plt.grid(True)
plt.show()
```

```python
# ===== SIMULATION SETTINGS =====
# These control how the simulation runs
dt = 0.1  # Time step [s]. Smaller is more accurate but slower.
total_time = 60.0  # Total simulation time [s]
time = np.arange(0, total_time, dt)  # Creates an array of time points [0, 0.1, 0.2, ... 59.9]

# ===== ROCKET PARAMETERS =====
# These are the properties of our rocket
g = 9.81  # Gravitational acceleration [m/s^2]
rho = 1.225  # Air density at sea level [kg/m^3]. Constant for simplicity.
Cd = 0.75  # Drag coefficient. A reasonable guess for a model rocket.
A = np.pi * (0.1)**2  # Cross-sectional area [m^2]. For a ~20cm diameter rocket.

# Initial Conditions
h0 = 0.0  # Initial altitude [m]
v0 = 0.0  # Initial velocity [m/s]
m0 = 20.0  # Initial mass (wet mass) [kg]

# Engine Properties
m_dot = 0.5  # Mass flow rate of propellant [kg/s]
burn_time = 10.0  # Engine burn time [s]
thrust = 400.0  # Constant thrust [N]. 400 N can lift a 20kg rocket.

# Print a summary to check our parameters
print("Simulation Parameters Set:")
print(f"Time step: {dt}s | Total time: {total_time}s")
print(f"Initial Mass: {m0} kg | Thrust: {thrust} N | Burn Time: {burn_time} s")
print(f"Drag Area: {A:.4f} m² | Drag Coeff: {Cd}")
```

```python
# ===== INITIALIZE ARRAYS TO STORE RESULTS =====
# We create arrays filled with zeros, one for each time step.
# We will fill these arrays in the loop below.
h = np.zeros(len(time))  # Altitude array [m]
v = np.zeros(len(time))  # Velocity array [m/s]
a = np.zeros(len(time))  # Acceleration array [m/s²]
m = np.zeros(len(time))  # Mass array [kg]

# Set the initial values (at time = 0)
h[0] = h0  # Initial altitude
v[0] = v0  # Initial velocity
m[0] = m0  # Initial mass
a[0] = (thrust - m[0]*g) / m[0]  # Initial acceleration ((T - mg)/m)

# ===== THE SIMULATION LOOP =====
# This is the core of the program. It calculates the state of the rocket at each new time step.
print("Starting simulation...")
for i in range(1, len(time)):
    # 1. Calculate current mass (if engine is still burning)
    if time[i] < burn_time:
        m[i] = m0 - m_dot * time[i]  # Mass decreases linearly
        current_thrust = thrust
    else:
        m[i] = m0 - m_dot * burn_time  # Mass remains constant after burn
        current_thrust = 0  # Engine cuts off

    # 2. Calculate DRAG force: D = 0.5 * density * v^2 * Cd * A
    # Use the velocity from the previous time step (v[i-1])
    Drag = 0.5 * rho * (v[i-1]**2) * Cd * A
    # Drag acts against the direction of motion.
    if v[i-1] > 0:
        Drag = -Drag  # If rocket is going up, drag points down.
    else:
        Drag = Drag   # If rocket is going down, drag points up.

    # 3. Calculate NET FORCE and ACCELERATION
    # F_net = Thrust + Weight + Drag
    # Weight is always downward: -m[i] * g
    F_net = current_thrust - m[i]*g + Drag
    a[i] = F_net / m[i]  # a = F_net / m

    # 4. INTEGRATE to find new velocity and height (Euler's Method)
    v[i] = v[i-1] + a[i] * dt
    h[i] = h[i-1] + v[i-1] * dt

print("Simulation complete!")
```

```python
# ===== PLOT THE RESULTS =====
# Create a figure with 2 rows and 2 columns of subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10)) # Makes a larger figure
fig.suptitle('Rocket Trajectory Simulation Results', fontsize=16, fontweight='bold')

# Plot 1: Altitude vs. Time
axs[0, 0].plot(time, h, 'b-', linewidth=2)
axs[0, 0].set_ylabel('Altitude [m]', fontweight='bold')
axs[0, 0].grid(True, linestyle='--', alpha=0.7)
axs[0, 0].set_title('Altitude')

# Plot 2: Velocity vs. Time
axs[0, 1].plot(time, v, 'r-', linewidth=2)
axs[0, 1].set_ylabel('Velocity [m/s]', fontweight='bold')
axs[0, 1].grid(True, linestyle='--', alpha=0.7)
axs[0, 1].set_title('Velocity')

# Plot 3: Acceleration vs. Time
axs[1, 0].plot(time, a, 'g-', linewidth=2)
axs[1, 0].set_ylabel('Acceleration [m/s²]', fontweight='bold')
axs[1, 0].set_xlabel('Time [s]', fontweight='bold')
axs[1, 0].grid(True, linestyle='--', alpha=0.7)
axs[1, 0].set_title('Acceleration')

# Plot 4: Mass vs. Time
axs[1, 1].plot(time, m, 'm-', linewidth=2)
axs[1, 1].set_ylabel('Mass [kg]', fontweight='bold')
axs[1, 1].set_xlabel('Time [s]', fontweight='bold')
axs[1, 1].grid(True, linestyle='--', alpha=0.7)
axs[1, 1].set_title('Mass')

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()

# ===== CALCULATE AND PRINT KEY PERFORMANCE METRICS =====
# Find the index of maximum altitude
max_alt_index = np.argmax(h)
max_alt = h[max_alt_index]
time_at_max_alt = time[max_alt_index]

# Find maximum velocity and acceleration
max_vel = np.max(v)
max_accel = np.max(a)

print("\n*** KEY FLIGHT RESULTS ***")
print(f"Maximum Altitude (Apogee): {max_alt:.2f} m at {time_at_max_alt:.2f} s")
print(f"Maximum Velocity: {max_vel:.2f} m/s")
print(f"Maximum Acceleration: {max_accel:.2f} m/s² ({max_accel / g:.2f} Gs)")
print(f"Burnout Mass: {m[-1]:.2f} kg") # m[-1] gets the last element in the mass array
```

```python
# ===== INITIALIZE ARRAYS (Now with Dynamic Pressure) =====
h = np.zeros(len(time))  # Altitude [m]
v = np.zeros(len(time))  # Velocity [m/s]
a = np.zeros(len(time))  # Acceleration [m/s²]
m = np.zeros(len(time))  # Mass [kg]
q = np.zeros(len(time))  # Dynamic Pressure [Pa]

# Set initial values
h[0] = h0
v[0] = v0
m[0] = m0
# Calculate initial acceleration and dynamic pressure
Drag_force = 0.5 * rho * (v[0]**2) * Cd * A * (-1 if v[0] >= 0 else 1) # Drag opposes motion
a[0] = (thrust - m[0]*g + Drag_force) / m[0]
q[0] = 0.5 * rho * v[0]**2 # Dynamic Pressure = 1/2 * density * velocity^2

# ===== UPGRADED SIMULATION LOOP =====
print("Starting simulation...")
for i in range(1, len(time)):
    # 1. Calculate current mass and thrust
    if time[i] < burn_time:
        m[i] = m0 - m_dot * time[i]
        current_thrust = thrust
    else:
        m[i] = m0 - m_dot * burn_time
        current_thrust = 0

    # 2. Calculate DRAG force using current mass and UPDATED velocity
    # The sign is set to oppose the direction of motion
    Drag_force = 0.5 * rho * (v[i-1]**2) * Cd * A
    if v[i-1] >= 0:
        Drag_force = -Drag_force  # If going up, drag points down
    # else, if velocity is negative (down), drag is positive (up)

    # 3. Calculate NET FORCE and ACCELERATION
    F_net = current_thrust - m[i]*g + Drag_force
    a[i] = F_net / m[i]

    # 4. INTEGRATE to find new velocity and height (Euler's Method)
    v[i] = v[i-1] + a[i] * dt
    h[i] = h[i-1] + v[i-1] * dt

    # 5. CALCULATE DYNAMIC PRESSURE at the current time step
    q[i] = 0.5 * rho * v[i]**2 # <-- This is a key performance metric!

print("Simulation complete!")
```

```python
# ===== COMPREHENSIVE PLOT =====
fig, axs = plt.subplots(3, 2, figsize=(16, 12)) # 3 rows, 2 columns
fig.suptitle('Comprehensive Rocket Trajectory Simulation Results', fontsize=16, fontweight='bold')

# Plot 1: Altitude
axs[0, 0].plot(time, h, 'b-', linewidth=2)
axs[0, 0].set_ylabel('Altitude [m]', fontweight='bold')
axs[0, 0].grid(True, linestyle='--', alpha=0.7)
axs[0, 0].set_title('Trajectory')

# Plot 2: Velocity
axs[0, 1].plot(time, v, 'r-', linewidth=2)
axs[0, 1].set_ylabel('Velocity [m/s]', fontweight='bold')
axs[0, 1].grid(True, linestyle='--', alpha=0.7)
axs[0, 1].set_title('Velocity')

# Plot 3: Acceleration
axs[1, 0].plot(time, a, 'g-', linewidth=2)
axs[1, 0].set_ylabel('Acceleration [m/s²]', fontweight='bold')
axs[1, 0].grid(True, linestyle='--', alpha=0.7)
axs[1, 0].set_title('Acceleration')

# Plot 4: Mass
axs[1, 1].plot(time, m, 'm-', linewidth=2)
axs[1, 1].set_ylabel('Mass [kg]', fontweight='bold')
axs[1, 1].set_xlabel('Time [s]', fontweight='bold')
axs[1, 1].grid(True, linestyle='--', alpha=0.7)
axs[1, 1].set_title('Mass')

# Plot 5: Dynamic Pressure
axs[2, 0].plot(time, q, 'k-', linewidth=2) # 'k-' is black line
axs[2, 0].set_ylabel('Dynamic Pressure [Pa]', fontweight='bold')
axs[2, 0].set_xlabel('Time [s]', fontweight='bold')
axs[2, 0].grid(True, linestyle='--', alpha=0.7)
axs[2, 0].set_title('Dynamic Pressure (q)')

# Hide the empty subplot (3rd row, 2nd column)
axs[2, 1].set_visible(False)

plt.tight_layout()
plt.show()

# ===== PRINT KEY RESULTS =====
max_alt_index = np.argmax(h)
max_alt = h[max_alt_index]
time_at_max_alt = time[max_alt_index]
max_q = np.max(q)
time_at_max_q = time[np.argmax(q)]

print("\n*** KEY FLIGHT RESULTS ***")
print(f"Maximum Altitude (Apogee): {max_alt:.2f} m at {time_at_max_alt:.2f} s")
print(f"Maximum Velocity: {np.max(v):.2f} m/s at {time[np.argmax(v)]:.2f} s")
print(f"Maximum Acceleration: {np.max(a):.2f} m/s² ({np.max(a) / g:.2f} Gs)")
print(f"Maximum Dynamic Pressure (Max Q): {max_q:.2f} Pa at {time_at_max_q:.2f} s")
print(f"Burnout Mass: {m[-1]:.2f} kg")
```

## Investigation 1: Burnout Acceleration Spike

### Observation
The results show a massive negative acceleration spike at engine burnout (t=10s).

### Hypothesis
The spike is caused by the simultaneous occurrence of two factors:
1.  **Sudden Loss of Thrust:** The engine cuts off, removing the large upward force.
2.  **Minimum Mass:** The rocket has burned all its fuel, so its mass is at the lowest point of the flight.

### Conclusion
With thrust gone, the only remaining forces are weight (down) and drag (down). The net force becomes large and negative. According to Newton's Second Law (a = F_net / m), this large negative net force acting on a very small mass results in the observed extreme negative acceleration.

```python
# ===== "WHAT IF" PARAMETERS =====
# We are creating a new rocket design to compare against the original

# Let's increase thrust by 25%
new_thrust = 1.25 * thrust  # 500 Newtons instead of 400

print("=== Original Rocket ===")
print(f"Thrust: {thrust} N")
print(f"Initial Mass: {m0} kg")
print(f"Burn Time: {burn_time} s")
print(f"Total Impulse: {thrust * burn_time} N·s")

print("\n=== New 'What If' Rocket ===")
print(f"Thrust: {new_thrust} N")
print(f"Initial Mass: {m0} kg") # We keep the mass the same
print(f"Burn Time: {burn_time} s") # We keep the burn time the same
print(f"Total Impulse: {new_thrust * burn_time} N·s")
```

```python
# ===== SIMULATE BOTH ROCKETS FOR COMPARISON =====

# --- Simulation function ---
# Let's define a function so we can easily simulate different rockets
def simulate_rocket(thrust_value, m0_value, burn_time_value):
    # Initialize arrays
    h = np.zeros(len(time))
    v = np.zeros(len(time))
    a = np.zeros(len(time))
    m = np.zeros(len(time))

    # Set initial values
    h[0] = h0
    v[0] = v0
    m[0] = m0_value
    Drag_force = 0.5 * rho * (v[0]**2) * Cd * A * (-1 if v[0] >= 0 else 1)
    a[0] = (thrust_value - m[0]*g + Drag_force) / m[0]

    # Simulation Loop
    for i in range(1, len(time)):
        # Mass and Thrust
        if time[i] < burn_time_value:
            m[i] = m0_value - m_dot * time[i]
            current_thrust = thrust_value
        else:
            m[i] = m0_value - m_dot * burn_time_value
            current_thrust = 0

        # Drag force
        Drag_force = 0.5 * rho * (v[i-1]**2) * Cd * A
        if v[i-1] >= 0:
            Drag_force = -Drag_force

        # Net Force and Acceleration
        F_net = current_thrust - m[i]*g + Drag_force
        a[i] = F_net / m[i]

        # Integrate
        v[i] = v[i-1] + a[i] * dt
        h[i] = h[i-1] + v[i-1] * dt

    return h, v, a, m

# --- Simulate Original Rocket ---
h_orig, v_orig, a_orig, m_orig = simulate_rocket(thrust, m0, burn_time)
# --- Simulate New "What If" Rocket ---
h_new, v_new, a_new, m_new = simulate_rocket(new_thrust, m0, burn_time)

# --- Plot Comparison Results ---
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('"What If" Analysis: 25% Increase in Thrust', fontsize=16, fontweight='bold')

# Plot 1: Altitude Comparison
axs[0, 0].plot(time, h_orig, 'b-', label='Original (400 N)', linewidth=2)
axs[0, 0].plot(time, h_new, 'r--', label='New (500 N)', linewidth=2)
axs[0, 0].set_ylabel('Altitude [m]', fontweight='bold')
axs[0, 0].legend()
axs[0, 0].grid(True, linestyle='--', alpha=0.7)
axs[0, 0].set_title('Trajectory: Altitude vs. Time')

# Plot 2: Velocity Comparison
axs[0, 1].plot(time, v_orig, 'b-', label='Original (400 N)', linewidth=2)
axs[0, 1].plot(time, v_new, 'r--', label='New (500 N)', linewidth=2)
axs[0, 1].set_ylabel('Velocity [m/s]', fontweight='bold')
axs[0, 1].legend()
axs[0, 1].grid(True, linestyle='--', alpha=0.7)
axs[0, 1].set_title('Velocity vs. Time')

# Plot 3: Acceleration Comparison
axs[1, 0].plot(time, a_orig, 'b-', label='Original (400 N)', linewidth=2)
axs[1, 0].plot(time, a_new, 'r--', label='New (500 N)', linewidth=2)
axs[1, 0].set_ylabel('Acceleration [m/s²]', fontweight='bold')
axs[1, 0].set_xlabel('Time [s]', fontweight='bold')
axs[1, 0].legend()
axs[1, 0].grid(True, linestyle='--', alpha=0.7)
axs[1, 0].set_title('Acceleration vs. Time')

# Plot 4: Mass (They are the same, so just plot one)
axs[1, 1].plot(time, m_orig, 'g-', label='Mass (Both Rockets)', linewidth=2)
axs[1, 1].set_ylabel('Mass [kg]', fontweight='bold')
axs[1, 1].set_xlabel('Time [s]', fontweight='bold')
axs[1, 1].legend()
axs[1, 1].grid(True, linestyle='--', alpha=0.7)
axs[1, 1].set_title('Mass vs. Time')

plt.tight_layout()
plt.show()

# --- Print Key Comparison Results ---
max_alt_orig = np.max(h_orig)
max_alt_new = np.max(h_new)
altitude_increase = ((max_alt_new - max_alt_orig) / max_alt_orig) * 100

max_vel_orig = np.max(v_orig)
max_vel_new = np.max(v_new)
velocity_increase = ((max_vel_new - max_vel_orig) / max_vel_orig) * 100

print("\n*** COMPARISON RESULTS ***")
print(f"Original Max Altitude: {max_alt_orig:.2f} m")
print(f"New Max Altitude:      {max_alt_new:.2f} m")
print(f"Altitude Increase:     {altitude_increase:.2f}%")
print("---")
print(f"Original Max Velocity: {max_vel_orig:.2f} m/s")
print(f"New Max Velocity:      {max_vel_new:.2f} m/s")
print(f"Velocity Increase:     {velocity_increase:.2f}%")
```

## Investigation 2: Thrust vs. Performance Trade-off

### Hypothesis
A 25% increase in thrust (from 400 N to 500 N) will result in a higher maximum velocity and apogee.

### Results
The simulation confirmed the hypothesis. The higher thrust design achieved a 40.21% increase in altitude and a 30.39% increase in velocity.

### Key Insight & Engineering Trade-off
The relationship was not linear. The altitude did not increase by 25%. This reveals a critical design trade-off:

-   **Pro:** Higher thrust yields better performance (altitude/velocity).
-   **Con:** The higher thrust caused a significantly more severe acceleration spike at burnout.

**Root Cause:** The increased thrust led to a much higher velocity. Since aerodynamic drag is proportional to the square of velocity (`D ∝ v²`), the drag force became disproportionately larger. This larger drag force, which opposes motion, resulted in a much larger negative net force the instant the thrust terminated at burnout.
